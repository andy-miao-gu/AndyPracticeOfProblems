````python
import math as mt
````

````python
def ILBsin(I, L, B, alpha):
    return I * L * B * mt.sin(alpha)
````

````python
electricity = 10 
magnetic_field = 0.5
length_of_wire = 5
angle_rod_with_field = 45

# find Force using ILBsin function
````

````python
ILBsin(electricity, length_of_wire, magnetic_field, angle_rod_with_field)
````
