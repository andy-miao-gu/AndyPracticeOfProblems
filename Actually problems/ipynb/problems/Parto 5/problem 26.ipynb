```python
import sympy as sp

# Define the variable and the function
x = sp.Symbol('x')
#f = x**2 - 4  # Example function f(x) = x^2 - 4
# x_der =2*x -88
f = x**2 - 88*x +86

# Derivative of the function
f_prime = sp.diff(f, x)

# Define the initial guess

x1=1
for i in range(10):
    x0 = x1

    # Evaluate the function and its derivative at the initial guess
    f_x0 = f.subs(x, x0)
    f_prime_x0 = f_prime.subs(x, x0)

    # Perform one iteration of the Newton-Raphson method
    x1 = x0 - f_x0 / f_prime_x0

    # Display the result
    print(f"Initial guess: {x0}")
    print(f"Function value at initial guess: {f_x0}")
    print(f"Derivative value at initial guess: {f_prime_x0}")
    print(f"Next approximation: {x1}")
```
